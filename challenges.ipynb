1. The "Seed" List
Since you cannot guess .onion URLs, your script needs a "Seed List." You can create a function that scrapes a directory (like The Hidden Wiki or TorLinks) to update your list of URL targets dynamically.

Challenge: 80% of links on these directories are dead at any given time.

Solution: Your script needs a "Link Validator" module that discards URLs that timeout 3 times in a row.

2. Rotation and Persistence
For 24/7 operation, you cannot rely on a single Tor circuit.

Circuit Rotation: Use stem (as mentioned previously) to request a new identity every ~100 requests. This prevents your scraper from getting "stuck" on a slow circuit.

Database: Do not print to console. Save matches to a local sqlite database or a .json log file so you can review alerts without staring at the screen.

3. Handling "Google" Alerts vs. Real Dumps
You mentioned Google found your info. Google usually indexes text that has leaked onto the clear web (e.g., a pastebin link that was shared on Reddit).

Real dark web monitoring involves downloading Data Dumps (SQL files, CSVs).

A web scraper is often insufficient for this. The "Pro" move here is not just to scrape HTML, but to look for file extensions like .zip, .sql, or .txt within the onion sites, download them, and then grep the files locally for your information.
