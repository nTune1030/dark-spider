1. Search Indexes (The "Google" of Tor): There are legal, indexable search engines on Tor (e.g., Ahmia). Your scraper should query these first to find which .onion sites currently contain your keywords.

2. Paste Sites: Similar to Pastebin, but on Tor. These are text-only dumps where data leaks are often posted. They are easier to scrape because they rarely require JavaScript or logins.

3. Breach Marketplaces: This is where most data actually lives, but it is extremely difficult to scrape programmatically. They require CAPTCHA solving, account creation, and 2FA. For a personal project, I recommend skipping these and focusing on public dumps.
